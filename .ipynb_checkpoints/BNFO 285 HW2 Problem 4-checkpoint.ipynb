<h3> Problem 4

```r
data=read.csv("hw2-p4-data.csv", header=FALSE, stringsAsFactors = FALSE)

africanCoordinates=data[(data[,3]=="African"),1:2]
europeanCoordinates=data[(data[,3]=="European"),1:2]

#Checking values by plotting
plot(africanCoordinates, col='red', main="African Coordinates")
plot(europeanCoordinates, col='green', main="European Coordinates")
```

The data looks roughly bivariate normally distributed. Bivariate Normal Distributions can be used to model the two distributions and calculate probability density. 
I will be Using Maximum Likelihood estimation to estimate the probability density functions for both the distributions. Then the following decision rule will be use to classify the unclassified points:
1. If the probability density for europeans distribution is higher then african for the unclassified point. Then the poiont will be classified as European.
2. If not then the point will be classified as African.


The maximum likelihood estimator for the means vector is the mean of the individual coordinates

```r
meansForAfricanCoordinates=apply(africanCoordinates,2,mean)
meansForAfricanCoordinates
```

```r
meansForEuropeanCoordinates=apply(europeanCoordinates,2,mean)
meansForEuropeanCoordinates
```

The maximum likelihood estimator for the Covariance is the : Covariance matrix * 
(n-1)/n .
Where n is the number of samples.

```r
nE=nrow(europeanCoordinates)
nE
```

```r
nA=nrow(africanCoordinates)
nA
```

Calculating Maximum Likelihood estimates of Covariance Matrices

```r
CovE=cov(europeanCoordinates)*((nE-1)/nE)
CovE
```

```r
CovA=cov(africanCoordinates)*((nA-1)/nA)
CovA
```

Getting European Probability Density Function

```r
euroPeanProbailityDensity= function(coordinate){
    V=t(coordinate-meansForEuropeanCoordinates)%*%solve(CovE)%*%(coordinate-meansForEuropeanCoordinates)
    probabilityDensity= (det(2*pi*CovE)^(-1/2))*exp((-1/2)*V)                 
}
```

Getting African Probability Density Function

```r
africanProbabilityDensity=function(coordinate){
    V=t(coordinate-meansForAfricanCoordinates)%*%solve(CovA)%*%(coordinate-meansForAfricanCoordinates)
    probabilityDensity=(det(2*pi*CovA)^(-1/2))*exp((-1/2)*V)
}
```

<h2>Getting Unclassified Coordinates

```r
africanLogicalVector=data[,3]=="African"
europeanLogicalVector=data[,3]=="European"
unclassifiedPointsLogicalVector=africanLogicalVector+europeanLogicalVector==0
unclassifiedCoordinates=data[unclassifiedPointsLogicalVector,1:2]
```

```r
europeanProbabilities=apply(unclassifiedCoordinates,1,euroPeanProbailityDensity)
africanProbabilities=apply(unclassifiedCoordinates,1,africanProbabilityDensity)
```

From the probabilities Matrix is seems that 10 points each clearly below to each distribution and 10 points are unclear

```r
ProbabilitiesMatrix=cbind(europeanProbabilities,africanProbabilities)
colnames(ProbabilitiesMatrix)=c("European Probabilites From Model","African Probabilities From Model")
Classification=apply(ProbabilitiesMatrix,1,function(vals){
    if(vals[1]>vals[2]){
        return("European")
    }
    else{
        return("African")
    }
})
cbind(ProbabilitiesMatrix,Classification)
```
